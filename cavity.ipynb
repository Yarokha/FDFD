```python
import numpy as np
import scipy as sp
import scipy.sparse.linalg as linalg
import matplotlib.pyplot as plt

a = 1
n_grid = 200
dz = a/n_grid
eps = 1
mu = 1

one = (np.ones(n_grid)/dz)
F = sp.sparse.dia_matrix(([-one,one], [0,1]), [n_grid, n_grid])
B = sp.sparse.dia_matrix(([-one,one], [-1,0]), [n_grid, n_grid])

EM = -1/eps/mu*F*B

kt = 2*sp.pi*1
num_modes = 6
k2, V = linalg.eigs(EM, k=num_modes, M=None, sigma=kt**2)

V_indicies = np.linspace(0,num_modes-1, num_modes, dtype=np.int)
k, V_indicies = (list(t) for t in zip(*sorted(zip(np.sqrt(k2), V_indicies))))
lam = 2*sp.pi/np.real(k)
plt.plot(np.linspace(1,num_modes, num_modes), lam, '-')
plt.plot(np.linspace(1,num_modes, num_modes), 4/(2*np.linspace(0,5, num_modes)+1), '*')  # Analytic
```

```python
plt.show()
```

```python
f, ax = plt.subplots(num_modes,1)
for n in np.arange(V.shape[1]):
    ax[n].plot((V[:,V_indicies[n]]),'-')
```
